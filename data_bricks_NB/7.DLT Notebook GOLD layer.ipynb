### Creating Expectaions

```python
looktables_rules = {
"rule1" : 'show_id is NOT NULL'
}
```

```python
@dlt.table(
    name = "gold_netflix_directors"
)
@dlt.expect_all_or_drop(looktables_rules)
def gold_netflix_directors():
    df = spark.readStream.format("delta").load('abfss://silver@mynetflxstrgacc.dfs.core.windows.net/netflix_directors')
    return df
```
